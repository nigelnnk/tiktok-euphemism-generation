```python
dict_word_to_code = dict()
for row in df_x.iterrows():
    combi_str = row[1]['Unnamed: 13']
    idx = combi_str.find(',')
    word = combi_str[0:idx]
    label = combi_str[idx + 1:]
    print(word)
    print(label)
    dict_word_to_code[word] = label
```

```python
dict_word_to_code['semen'] = 'A'
dict_word_to_code['rapist'] = 'V'
dict_word_to_code['raping'] = 'V'
```

```python
original_code = []
for row in df_y.iterrows():
    word = row[1]['Original']
    try:
        original_code.append(dict_word_to_code[word])
    except:
        print(word)
```

```python
df_y['orig_code'] = original_code
```

```python
from Levenshtein import distance as lev
import numpy as np
```

```python
# Count of euphemisms that are in same category
count_same_cat = 0
# Count of euphemisms that have similar meaning (>= 4)
count_same_meaning = 0
# Close Euph
good_euph = []
# % edit distance of euphemisms compared to baselines
edit_distances = []
for row in df_y.iterrows():
    orig = row[1]['Original']
    euph = row[1]['Generated Slang']
    if row[1]['Tiebreak'] == row[1]['orig_code'] and row[1]['Tiebreak Sim'] > 4:
        good_euph.append(euph)
    if row[1]['Tiebreak'] == row[1]['orig_code']:
        count_same_cat += 1
    if row[1]['Tiebreak Sim'] > 4:
        count_same_meaning += 1
    edit_distances.append(lev(orig, euph))
```

```python
# Average Similarity
np.average(df_y['Tiebreak Sim'])
```

```python
# Average Edit Distance
np.average(edit_distances)
```

```python
# % of euphemisms that are in same category
print(count_same_cat / float(len(df_y.axes[0])))
# % of euphemisms that have similar meaning (>= 4)
print(count_same_meaning / float(len(df_y.axes[0])))
```

```python
# Good Euphemisms (same category as original word, > 4 similarity)
rgood_euph = set(good_euph) - set(dict_word_to_code.keys())
```

```python
curr_word = 'nog'
check = True
for euph in unbanned_good_euph:
    print(euph)
    if check and euph != curr_word:
        continue
    else:
        check = False
        
    curr_folder = f'./vids_to_analyze/{euph}'
    df_meta = pd.read_csv(os.path.join(curr_folder, f'tiktok.com_tag_{euph}.csv'))
    tiktok_videos = []
    for iid in list(df_meta['video_id']):
        tiktok_videos.append(f'https://www.tiktok.com/@tiktok/video/{iid}?is_copy_url=1&is_from_webapp=v1')
    links = pykk.save_tiktok_multi_urls(tiktok_videos, save_video=True, save_video_link = False, sleep = 15)
    
    
    for file in os.listdir('./'):
        if '.mp4' in file or '_tag_' in file:
            shutil.move(os.path.join('./', file), os.path.join(curr_folder, file))
    
```

```python
df_annotate_all = []
for euph in unbanned_good_euph:
    print(euph)
    curr_folder = f'./vids_to_analyze/{euph}'
    df_meta = pd.read_csv(os.path.join(curr_folder, f'tiktok.com_tag_{euph}.csv'))
    df_meta.sort_values(by=['video_playcount'], ascending=False)
    count = 0
    for idx, row in enumerate(df_meta.iterrows()):
        #print(row[1])
        idd = row[1]['video_id']
        video_file = f'tiktok_video_{idd}.mp4'
        
        if count > 9:
            break
        # Video not available
        try:
            if os.path.getsize(os.path.join(curr_folder, video_file)) < 1000:
                continue
        except:
            continue
            
        df_annotate_all.append({'Video ID': row[1]['video_id'], 'Author': row[1]['author_username'], 'URL': row[1]['url_video'],
                                'Video Description': row[1]['video_description'], 'Video #Views': row[1]['video_playcount'],
                                'Euphemism': euph, 'Annotator 1': '', 'Letter 1': '', 'Annotator 2': '', 'Letter 2': ''})
        count += 1
        
    
pd.DataFrame(df_annotate_all).to_csv('./consim_videoeval.csv')
```
