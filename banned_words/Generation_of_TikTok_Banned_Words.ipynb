## Implementation

### Load resources
Download list of bad words from CMU professor and Github, and initialise TikTokAPI scraper

```python
import pandas as pd
import os
import requests
import string

from TikTokApi import TikTokApi
import nest_asyncio
nest_asyncio.apply()

api = TikTokApi(custom_verify_fp='')

cmu_list = requests.get("https://www.cs.cmu.edu/~biglou/resources/bad-words.txt")
github_list = requests.get("https://raw.githubusercontent.com/LDNOOBW/List-of-Dirty-Naughty-Obscene-and-Otherwise-Bad-Words/master/en")

tiktok_tos = ['ana', 'anorexia', 'anorexianervosa', 'bonespo', 'bulimia', 'eatingdisorder', 'mia', 'proana', 'proanorexia', 'probulimia', 'promia', 'secretsociety', 'skinny', 'thighgap', 'thin', 'thinspiration', 'thinspo', 'binge eating', 'bulimianervosa', 'purgingdisorder', 'ed', 'self harm', 'self hatred', 'scars', 'suicide', 'unalive', 'depression', 'grooming', 'csam', 'pedophilia', 'sexcamming', 'sexcam', 'doxxing', 'blackmail', 'swatting', 'firearm', 'firearms', 'ammunition', 'drugs', 'alcohol', 'tobacco', 'vaping', 'nicotine', 'phishing', 'gambling', 'casino', 'betting', 'poker', 'lotteries', 'dismember', 'mutilate', 'gore', 'sadistic']
```

Process list of words by removing spaces and ensuring that only ascii characters are allowed as keywords

```python
words_to_test = []
words_to_test.extend(cmu_list.text.split('\n'))
words_to_test.extend(github_list.text.split('\n'))
words_to_test.extend(tiktok_tos)

printable = set(string.printable)
words_to_test = filter(lambda x: x.isascii() and len(x) > 0, map(lambda x: x.replace(' ', ''), words_to_test))
words_to_test = sorted(list(set(words_to_test)))
with open("words_to_test.csv", "w") as f:
    f.write("\n".join(words_to_test))
print(words_to_test)
```

Try keywords on TikTok

```python
df_construct_arr = []
banned_keywords = []
for keyword in words_to_test:
    try:
        # info_full() is not used as this is mainly for logging purposes
        df_construct_arr.append(api.hashtag(keyword).info())
    except:
        print(f"{keyword} is a banned keyword")
        banned_keywords.append(keyword)

results = pd.DataFrame(df_construct_arr)
results["videoCount"] = results.apply(lambda row: row["stats"]["videoCount"], axis=1)
results["viewCount"] = results.apply(lambda row: row["stats"]["viewCount"], axis=1)
results = results.drop("stats", axis=1)
results.to_csv('nonbanned_keywords_info.csv')

with open("banned_keywords.csv", "w") as f:
    f.write("\n".join(banned_keywords))
```

Also because a hashtag better represents a topic rather than general search which can take into account usernames or other context in videos
