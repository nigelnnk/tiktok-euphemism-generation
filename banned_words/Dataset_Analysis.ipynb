## TF-IDF on TikTok's Community Guidelines

As mentioned in https://www.tiktok.com/community-guidelines, we get the text of each section, use TF-IDF and analyse how similar the sections are in terms of cosine similarity. 

```python
from bs4 import BeautifulSoup
import requests

html_doc = requests.get("https://www.tiktok.com/community-guidelines?lang=en")
soup = BeautifulSoup(html_doc.text, 'html.parser')

headers = [h2.text for h2 in soup.find_all('h2')][1:-3]
```

```python
section_text = []
temp = []
for p_tag in soup.select("p, ul"):
    p_text = p_tag.text
    p_text = p_text.replace(u'\xa0', u' ')
    if len(p_text) == 0:
        section_text.append("\n".join(temp))
        temp.clear()
    elif "Do not" in p_text:
        for bullet in p_tag.next_sibling.find_all("li"):
            li_text = bullet.text
            temp.append(li_text)
    else:
        temp.append(p_text)

section_text.append("\n".join(temp))
temp.clear()

section_text = section_text[2:-5]
# print("\n---\n".join(section_text))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import normalize
import numpy as np

vectoriser = TfidfVectorizer()
matrix = vectoriser.fit_transform(section_text)
idxs = np.argsort(["M","D","S","A","B","H","E","N","I","V"])
matrix = matrix[idxs]

ans = []
for x in range(len(section_text)):
    temp = []
    for y in range(len(section_text)):
        temp.append(cosine_similarity(matrix[x], matrix[y]))
    ans.append(temp)
cm = np.asarray(ans).squeeze()

cm_display = ConfusionMatrixDisplay(cm, display_labels=labels)
fig, ax = plt.subplots(figsize=(8,8))
cm_display.plot(ax=ax, values_format=".2f")
plt.savefig("tfidf.png")
```

## Banned words analysis

```python
banned_words = []
with open("banned_keywords.csv", "r") as f:
    for line in f:
        banned_words.append(line.strip())

label_mapping = {}
with open("final_labels.csv", "r") as f:
    for line in f:
        temp = line.split(",")
        label_mapping[temp[0]] = temp[1].strip()

pred = []
true = []
count = 0
for k, v in label_mapping.items():
    true.append(True if k in banned_words else False)
    pred.append(True if v != "Z" else False)

# print(true)
# print(pred)

print(f"Number of positive labels: Pred {sum(pred)}\tTrue {sum(true)}")

label_cm = ConfusionMatrixDisplay.from_predictions(true, pred,
                                                   labels=[True,False],
                                                   normalize="all",
                                                   values_format=".2f")
print("\n")

fig, ax = plt.subplots()
label_cm.plot(ax=ax, values_format=".2f")
plt.savefig("label_cm.png")
```

```python
ban_labels = []
all_labels = []
for k, v in label_mapping.items():
    if k in banned_words:
        ban_labels.append(v)
    all_labels.append(v)

unique_labels = sorted(list(set(ban_labels)))
temp = ["Label,Banned,All"]
for ell in unique_labels:
    # temp.append(f"{ell},{ban_labels.count(ell)/len(ban_labels):.3f},{all_labels.count(ell)/len(all_labels):.3f}")
    temp.append(f"{ell},{ban_labels.count(ell)},{all_labels.count(ell)}")

print("\n".join(temp))
```
